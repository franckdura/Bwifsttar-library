# Bridge weigh-in-motion

Le modle BWIM affirme que l'effet induit par le dplacement d'un vehicule sur une structure suspendue peut s'exprimer comme suit:

$$ y(t) = \sum_{a=1}^{A} w_a h(t - \tau_a) $$

o 
 - $A$ - nombre d'essieux du vehicule,
 - $w_a$ - poid du vehicule sur l'essieu $a$,
 - $\tau_a$ - instant de passage de l'essieu $a$,
 - $h(t)$ - fonction qui decrit l'effet induit par un chargement unitaire qui se deplace sur le pont  la meme vitesse du vehicule consider.

## Discrtisation

Soit $f_s = 1/T_s$ la frequence d'echantillonage du systeme, et $t_0$ l'instant d'entre du vehicule sur le pont.

- Le signal observ est represent par un vecteur de taille $K$:

$$ {\bf y} = \big[y(t_0 + kT_s)\big]_{k\in\{0,\dots,K-1\}}. $$

- La ligne d'influence est represent par un vecteur de taille $M\le K$:

$$ {\bf h} = \big[h(mT_s)\big]_{m\in\{0,\dots,M-1\}}. $$

- L'instant $\tau_a$ est represent par un vecteur de taille $K$ qui prend la valeur $1$ en position $c_a$ (echantillon correspondant au temps $\tau_a$) et $0$ autrement:

$$
\boldsymbol{\delta}_a = \big[\delta_{k,c_a}\big]_{k\in\{0,\dots,K-1\}} = \big[0\;\dots\;0\;\underbrace{1}_{c_a}\;0\;\dots\;0\big]^\top.
$$

Le modle BWIM peut alors se traduire par la rlation suivante:

$$ {\bf y} = \sum_{a=1}^A w_a \, \boldsymbol{\delta}_{a} * {\bf h}, $$

o le symbole $*$ denote le produit de convolution. De plus, on remarque que la convolution par le vecteur de decalage $\boldsymbol{\delta}_{a}$ est equivalent au produit par une matrice de Toeplitz $\mathbf{D_a} \in \mathbb{R}^{K\times M}$. Le modle BWIM se reconduit alors au systeme lineaire

$$ {\bf y} = \sum_{a=1}^A w_a \mathbf{D_a} {\bf h}. $$

## Calibration

La calibration consiste  retrouver la ligne d'influence ${\bf h}$  partir du signal observ ${\bf y}$, des instants de passage $\tau_{1},\dots,\tau_{A}$, et des poids $w_1,\dots,w_a$ connus  l'avance. Pour ce faire, on rcrit le modle BWIM sous forme d'un systeme lineaire qui depend de ${\bf h}$, c'est  dire

$$ {\bf y} = {\bf T}{\bf h} \qquad\textrm{with}\quad {\bf T} = \sum_{a=1}^A w_a \mathbf{D_a}. $$

Si ${\bf T}$ est une matrice de rang plein (ce qui est possible car $M\le K$), on peut retrouver la ligne d'influence par la mthode des moindres carrs, qui consiste 

$$ \operatorname*{minimize}_{{\bf h} \in \mathbb{R}^M} \; \| {\bf T}{\bf h} - {\bf y} \|^2, $$

dont la solution est $ {\bf h}^* = ({\bf T}^\top{\bf T})^{-1} {\bf T}^\top{\bf y}$.

```python
observed = events[0]

influence = calibration(truck, observed, weights, length=701)

reconstructed, rescaled_influence = reconstruction(truck, weights, influence)
```

```python
time, speed, peaks = truck
meters = speed * time

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(meters[:len(rescaled_influence)], rescaled_influence)
plt.xlabel('Meters')
plt.subplot(1,2,2)
plt.plot(meters, reconstructed, linewidth=2, label='Reconstructed')
plt.plot(meters, observed, label='Observed', alpha=0.7)
plt.xlabel('Meters')
plt.legend()
plt.show()
```

## Pese en marche

La pese en marche consiste  retrouver les poids $w_1,\dots,w_a$  partir du signal observ ${\bf y}$, des instants de passage $\tau_{1},\dots,\tau_{A}$, et de la ligne d'influence ${\bf h}$ connue  l'avance. Pour ce faire, on rcrit le modle BWIM sous forme d'un systeme lineaire qui depend de ${\bf w} = [w_1,\dots,w_a]^\top$, c'est  dire

$$ {\bf y} = {\bf H}{\bf w} \qquad\textrm{with}\quad {\bf H} = \begin{bmatrix} \mathbf{D_1} {\bf h} & \mathbf{D_2} {\bf h} &\dots& \mathbf{D_A} {\bf h} \end{bmatrix}. $$

Si ${\bf H}$ est une matrice de rang plein, on peut retrouver les poids par la mthode des moindres carrs, qui consiste 

$$ \operatorname*{minimize}_{{\bf w} \in \mathbb{R}^A} \; \| {\bf H}{\bf w} - {\bf y} \|^2, $$

dont la solution est ${\bf w}^* = ({\bf H}^\top{\bf H})^{-1} {\bf H}^\top{\bf y}$.
